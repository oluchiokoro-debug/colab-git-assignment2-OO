```python
# Create an empty list
library = []

# Add books
library.append("Things Fall Apart")
library.append("The Great Gatsby")
library.append("To Kill a Mockingbird")
library.append("1984")

print("Books in the library:")
print(library)
```

```python
print("Library Catalogue:")

for index, book in enumerate(library, start=1):
    print(f"{index}. {book}")
```

```python
def check_book(book_name):
    if book_name in library:
        return f"{book_name} is available."
    else:
        return f"{book_name} is not available."

# Test
print(check_book("1984"))
print(check_book("Harry Potter"))
```

```python
with open("books.txt", "w") as file:
    file.write("Python Basics\n")
    file.write("Machine Learning\n")
    file.write("Artificial Intelligence\n")
```

```python
def read_books(filename):
    try:
        with open(filename, "r") as file:
            books = file.readlines()

        print("Books from file:")
        for book in books:
            print(book.strip())

    except FileNotFoundError:
        print("File not found.")

# Test
read_books("books.txt")
```

```python
import pickle
```

```python
books = ["Python", "AI", "Data Science"]

with open("books.pkl", "wb") as file:
    pickle.dump(books, file)
```

```python
def load_books(filename):
    try:
        with open(filename, "rb") as file:
            books = pickle.load(file)

        print("Books loaded:")
        print(books)

    except FileNotFoundError:
        print("Pickle file not found.")

# Test
load_books("books.pkl")
```

```python
import sqlite3
```

```python
conn = sqlite3.connect("library.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS books (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT
)
""")

conn.commit()
```

```python
def add_book_db(title):
    try:
        cursor.execute(
            "INSERT INTO books (title) VALUES (?)",
            (title,)
        )
        conn.commit()
        print(f"{title} added successfully.")

    except sqlite3.Error as e:
        print("Database error:", e)

# Test
add_book_db("Deep Learning")
add_book_db("Python Programming")
```

```python
cursor.execute("SELECT * FROM books")

for row in cursor.fetchall():
    print(row)
```

```python
conn.close()
```

```python
def remove_book(book):
    if book in library:
        library.remove(book)
        print(f"{book} removed.")
    else:
        print("Book not found.")

remove_book("1984")
```

```python
def update_book(old, new):
    if old in library:
        index = library.index(old)
        library[index] = new
        print("Book updated.")
    else:
        print("Book not found.")

update_book("The Great Gatsby", "The Great Gatsby Revised")
```
